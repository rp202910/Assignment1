```python
import matplotlib.pyplot as plt
```

```python
# Line 
x= [1, 3, 5, 7, 9, 11, 13]
y= [20, 30, 50, 45, 89, 12, 40]
plt.plot(x, y)
plt.show()
```

```python
x= [1, 3, 5, 7, 9, 11, 13]
y= [20, 30, 50, 45, 89, 12, 40]
plt.figure(figsize=(8,5))
plt.plot(x,y,lw=3,color="red",linestyle="dashed",label="Line graph")
plt.title("Line graph",fontsize=14,color='blue')
plt.xlabel("x-axis",fontsize=14,color='blue')
plt.ylabel("Y-axis",fontsize=14,color='blue')
plt.grid()
plt.legend(loc="best")#for position of label.2.best
plt.show()
```

```python
x= [1, 3, 5, 7, 9, 11, 13]
y= [20, 30, 50, 45, 89, 12, 40]
plt.figure(figsize=(8,5))
plt.title("Line graph", fontsize= 14, color='blue')
plt.xlabel("X Axis",  fontsize= 14, color='blue')
plt.ylabel("Y Axis",  fontsize= 14, color='blue')
plt.grid()
plt.scatter(x,y, lw=3, color="red", linestyle="dashed", label="Scatter graph",s=120,marker="*")
plt.legend(loc="best")
plt.show()
```

```python
x= [1, 3, 5, 7, 9, 11, 13]
y= [20, 30, 50, 45, 89, 12, 40]
plt.figure(figsize=(8,5))
plt.title("Line graph", fontsize= 14, color='blue')
plt.xlabel("X Axis",  fontsize= 14, color='blue')
plt.ylabel("Y Axis",  fontsize= 14, color='blue')
plt.grid()
plt.bar(x,y, lw=3, color=["red","blue"],  label="Bar graph",width=0.5)
plt.legend(loc="best")
plt.show()
```

```python
x= [1, 3, 5, 7, 9, 11, 13]
y= [20, 30, 50, 45, 89, 12, 40]
plt.figure(figsize=(8,5))
plt.title("Line graph", fontsize= 14, color='blue')
plt.xlabel("Y Axis",  fontsize= 14, color='blue')
plt.ylabel("x Axis",  fontsize= 14, color='blue')
plt.grid()
plt.barh(x,y, lw=3, color=["red","blue"],  label="Bar graph",height=0.5)
plt.legend(loc="best")
plt.show()
```
